```python
import numpy as np
import tensorflow as tf

import autokeras as ak
import os
import cv2

import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

# 데이터 불러오기 및 나누기
[train_test_split] <br>
test_size: 검증 셋 크기<br>
stratify: class별 비율 유지<br>
shuffle: 셔플<br>
random_state: 다른 환경에서 동일 결과 보장 <br>

```python
X = np.load('/content/drive/MyDrive/Colab Notebooks/img_data_X.npy')
Y = np.load('/content/drive/MyDrive/Colab Notebooks/img_data_Y.npy')
print(X.shape, Y.shape)

train_x, valid_x, train_y, valid_y = train_test_split(X, Y, test_size = 0.1, stratify=Y, shuffle=True, random_state=42)
```

# 모델 학습 및 평가
Autokeras의 ImageClassifier를 이용해 간단하게 사용 가능<br>
[ImageClassifier Parameters] <br>
num_classes: 클래스 수 <br>
max_trials: 시도해볼 모델 수 <br>
loss: 케라스 로스 함수 적용 가능 (Defalut: binary_crossentropy' or 'categorical_crossentropy) <br>
metrics: 케라스 메트릭 함수 적용 가능 (Defalut: 'accuracy')
overwrite: 이전 프로젝트 이어서 사용(: False) <br>
<br>
[fit Parameters] <br>
epochs: 에포크 설정 <br>
callbacks: 케라스 콜벡함수 리스트 <br>
validation_split: 훈련데이터에서 일정 비율 검증데이터로 사용 (Defalut = 0.2) <br>
validation_data: 검증 데이터 셋 설정 <br>

```python
clf = ak.ImageClassifier(max_trials=3, directory='/content/drive/MyDrive/Colab Notebooks')
clf.fit(train_x, train_y, epochs=50)
print('Accuracy: {}'.format(clf.evaluate(valid_x, valid_y)))

best_model = clf.export_model()
best_model.save('/content/drive/MyDrive/Colab Notebooks/Best_model.h5')
```
